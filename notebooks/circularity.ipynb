# Step 3: Run the Analysis

Run the two cells below to run circularity analysis on all your amphoras.

````python
import csv, os, sys
sys.path.insert(0,'../../code')
import circ
from itertools import izip
import numpy as np
import matplotlib.pyplot as plt
````

````python
circularities = []
for model in models:
    print('Model: %s' % model)
    im_path = './circles/circle_%s.png' % model
    bin_im = circ.get_bin_im(im_path)
    circularity = circ.get_circularity(bin_im)
    print('Circularity: %s \n\n----------------------------------\n\n' % circularity)
    circularities.append(circularity)
````

# Step 4: Visualize the results

Adjust the parameters in the cell below to customize the look of your figure. Run the cell to set the parameters.

*Tip: https://colorbrewer2.org/#type=sequential&scheme=BuGn&n=3 is a great place to look for colors that are colorblind safe, print friendly, and photocopy safe.*

````python
width = 0.001
min_circularity = 0.975
max_circularity = 0.99
min_num_amphoras = 0
max_num_amphoras = 20
interval_num_amphoras = 2
figure_width = 8
figure_height = 5
bar_color = '#DE956C'
resolution_dpi = 600     # set resolution of saved figures in dots per inch
````

Run the the cell below to plot the results, using the parameters set in the cell above.

````python
plt.figure(figsize=(figure_width, figure_height))
plt.hist(circularities, bins=np.arange(min_circularity, max_circularity, width)+width/2, color=bar_color)
plt.xticks(np.arange(min_circularity, max_circularity, width), rotation=45)
plt.yticks(np.arange(min_num_amphoras, max_num_amphoras, interval_num_amphoras))
plt.xlabel('\nCircularity = 4 * pi * area / (perimeter^s)')
plt.ylabel('Number of Amphoras\n')
plt.title('Circularities for Yassiada Amphoras, n = %s' % len(models))
plt.show()
````

# Step 5: Save the results

Run the cell below to create an empty folder called "results".

````python
if not os.path.exists('results'):
    os.mkdir('results')
````

Run the cell below if you want to save the circularity data for each model in a comma-separated value (CSV) file.

````python
with open('./results/circularities.csv', 'wb') as f:
    writer = csv.writer(f)
    writer.writerows(izip(models, circularities))
````

Run the cell below if you want to save the plot as a PNG.

````python
plt.savefig('./results/circularities_histogram.png', dpi=resolution_dpi)
````
